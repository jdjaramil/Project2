```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 14})
from matplotlib.dates import DateFormatter
```

```python
columns = ['positive', 'negative', 'compound', 'polarity', 'subjectivity', 'pos', 'neg', 'neu','neutral',
          'BERT_sentiment_POS', 'BERT_sentiment_NEG', 'BERT_sentiment_NEU']
```

```python
fet = pd.read_csv('fettermanClassified2.csv', parse_dates = ['date'])
temp = fet['BERT_sentiment'].copy()
fet = pd.get_dummies(fet, columns=['BERT_sentiment'])
fet['BERT_sentiment'] = temp
print(fet.shape)

oz = pd.read_csv('OzClassified2.csv', parse_dates = ['date'])
temp = oz['BERT_sentiment'].copy()
oz = pd.get_dummies(oz, columns=['BERT_sentiment'])
oz['BERT_sentiment'] = temp
print(oz.shape)
```

```python
print(oz.columns)
oz.head()
```

```python
print(fet.columns)
fet.head()
```

```python
'''
oz_input = Input dataset related to Oz
fetterman_input = Input dataset related to Fetterman
weighted_mean :  It is a boolean that states if when we group by on a daily basi

'''
```

```python
def plot_trend(oz_input, fet_input, weighted_mean = False, weight_by = 'likeCount', BERT = False ):
    
    if weighted_mean: 
        trend_fet = fet_input.groupby('date').apply(lambda x: pd.Series(np.average(x[columns], weights = x[weight_by], axis = 0), columns))
        trend_oz = oz_input.groupby('date').apply(lambda x: pd.Series(np.average(x[columns], weights = x[weight_by], axis = 0), columns))
    
    else:    
        trend_fet = fet_input.groupby('date')[columns].mean()
        trend_oz = oz_input.groupby('date')[columns].mean()
        
    fig, ax = plt.subplots(1, 2, figsize = (15, 6))
    
    if (BERT == False):
        POSITIVE = 'positive'
        NEGATIVE = 'negative'
        NEUTRAL = 'neutral'
    else:
        POSITIVE = 'BERT_sentiment_POS'
        NEGATIVE = 'BERT_sentiment_NEG'
        NEUTRAL = 'BERT_sentiment_NEU'
        
    


    color_map_democrats = ["#0015BC", "#0015BC", "#0015BC"]
    ax[0].stackplot(trend_fet.index, trend_fet.loc[:, POSITIVE], trend_fet.loc[:, NEUTRAL], trend_fet.loc[:, NEGATIVE], 
                  labels = ['Positive', 'Neutral', 'Negative'],
                colors = color_map_democrats)
    ax[0].plot(trend_fet.index, trend_fet.loc[:, POSITIVE], color = 'white')
    ax[0].plot(trend_fet.index, trend_fet.loc[:, POSITIVE] + trend_fet.loc[:, NEUTRAL], color = 'white')

    children = ax[0].get_children()    
    children[0].set_alpha(1)
    children[1].set_alpha(0.6)
    children[2].set_alpha(0.2)

    ax[0].set_title('Democrats Tweets Trend over time (Fetterman)\n')
    ax[0].set_xlabel('\nTime')
    ax[0].set_ylabel('\nProportion of tweets')
    ax[0].legend()


    color_map_republicans = ["#DE0100", "#DE0100", "#DE0100"]
    ax[1].stackplot(trend_oz.index, trend_oz.loc[:, POSITIVE], trend_oz.loc[:, NEUTRAL], trend_oz.loc[:, NEGATIVE], 
                  labels = ['Positive', 'Neutral', 'Negative'],
                colors = color_map_republicans)
    ax[1].plot(trend_oz.index, trend_oz.loc[:, POSITIVE], color = 'white')
    ax[1].plot(trend_oz.index, trend_oz.loc[:, POSITIVE] + trend_oz.loc[:, NEUTRAL], color = 'white')

    children = ax[1].get_children()    
    children[0].set_alpha(1)
    children[1].set_alpha(0.6)
    children[2].set_alpha(0.2)

    ax[1].set_title('Republicans Tweets Trend over time (Dr. Oz)\n')
    ax[1].set_xlabel('\nTime')
    ax[1].set_ylabel('\nProportion of tweets')
    ax[1].legend()
    myFmt = DateFormatter("%m/%d")
    ax[0].xaxis.set_major_formatter(myFmt)
    ax[1].xaxis.set_major_formatter(myFmt)
    fig.autofmt_xdate()
    plt.show()

    if (BERT == False):
        fig, ax = plt.subplots(1, 2, figsize = (16, 5))

        ax[0].plot(trend_fet.index, trend_fet.loc[:, 'polarity'], label = 'Fetterman', color = color_map_democrats[0])
        ax[0].plot(trend_oz.index, trend_oz.loc[:, 'polarity'], label = 'Dr. Oz', color = color_map_republicans[0])
        ax[0].grid()
        ax[0].set_xlabel('Time')
        ax[0].set_ylabel('Polarity')
        ax[0].set_title('Polarity')
        ax[0].legend()

        ax[1].plot(trend_fet.index, trend_fet.loc[:, 'compound'], label = 'Fetterman', color = color_map_democrats[0])
        ax[1].plot(trend_oz.index, trend_oz.loc[:, 'compound'], label = 'Dr. Oz', color = color_map_republicans[0])
        ax[1].grid()
        ax[1].set_xlabel('Time')
        ax[1].set_ylabel('Compound')
        ax[1].set_title('Compound')
        ax[1].legend()

        myFmt = DateFormatter("%m/%d")
        ax[0].xaxis.set_major_formatter(myFmt)
        ax[1].xaxis.set_major_formatter(myFmt)
        fig.autofmt_xdate()
        plt.show()

    fig, ax = plt.subplots(1, 2, figsize = (16, 5))

    ax[0].plot(trend_fet.index, trend_fet.loc[:, POSITIVE], label = 'Fetterman', color = color_map_democrats[0])
    ax[0].plot(trend_oz.index, trend_oz.loc[:, POSITIVE], label = 'Dr. Oz', color = color_map_republicans[0])
    ax[0].grid()
    ax[0].set_xlabel('Time')
    ax[0].set_ylabel('Positive Sentiment')
    ax[0].set_title('Positive Sentiment')
    ax[0].legend()

    ax[1].plot(trend_fet.index, trend_fet.loc[:, NEGATIVE], label = 'Fetterman', color = color_map_democrats[0])
    ax[1].plot(trend_oz.index, trend_oz.loc[:, NEGATIVE], label = 'Dr. Oz', color = color_map_republicans[0])
    ax[1].grid()
    ax[1].set_xlabel('Time')
    ax[1].set_ylabel('Negative Sentment')
    ax[1].set_title('Negative Sentiment')
    ax[1].legend()
    
    myFmt = DateFormatter("%m/%d")
    ax[0].xaxis.set_major_formatter(myFmt)
    ax[1].xaxis.set_major_formatter(myFmt)
    fig.autofmt_xdate()
    plt.show()
    
    fig, ax = plt.subplots(1, 2, figsize = (15, 6))

    ax[0].plot(trend_fet.index, trend_fet.loc[:, POSITIVE]-trend_fet.loc[:, NEGATIVE], label = 'Fetterman', color = color_map_democrats[0])
    ax[0].plot(trend_oz.index, trend_oz.loc[:, POSITIVE]- trend_oz.loc[:, NEGATIVE], label = 'Dr. Oz', color = color_map_republicans[0])
    ax[0].grid()
    ax[0].set_xlabel('Time')
    ax[0].set_ylabel('(Positive - Negative) Sentiment')
    ax[0].set_title('(Positive - Negative) Sentiment')
    ax[0].legend()
    myFmt = DateFormatter("%m/%d")
    ax[0].xaxis.set_major_formatter(myFmt)
    
    
    ax[1].hist(trend_fet[POSITIVE] - trend_fet[NEGATIVE], bins = 20, alpha = 0.9, color = color_map_democrats[0], label = 'Fetterman')
    ax[1].hist(trend_oz[POSITIVE] - trend_oz[NEGATIVE], bins = 20, alpha = 0.9, color = color_map_republicans[0], label = 'Oz')
    ax[1].legend()
    ax[1].set_title('(Positive - Negative) Sentiment')
    ax[1].grid()
    
    fig.autofmt_xdate()
    plt.show()
    
    
    

```

# All data: no filter

```python
plot_trend(oz, fet)
```

```python
plot_trend(oz, fet , BERT = True)
```

# Filter tweets with exclamation

```python
plot_trend(oz[(oz["content"].str.contains('!'))], fet[(fet["content"].str.contains('!'))]) 
```

```python
plot_trend(oz[(oz["content"].str.contains('!'))], fet[(fet["content"].str.contains('!'))], BERT = True) 
```

# Filter by most followers

```python
fetMostFoll = fet[fet['followersCount'] > fet[['followersCount']].quantile(q = 0.95).values[0]]
ozMostFoll = oz[oz['followersCount'] > oz[['followersCount']].quantile(q = 0.95).values[0]]
plot_trend(ozMostFoll, fetMostFoll)
```

```python
plot_trend(ozMostFoll, fetMostFoll, BERT = True)
```

# Filter by most likes

```python
fetMostLikes = fet[fet['likeCount'] > fet[['likeCount']].quantile(q = 0.95).values[0]]
ozMostLikes = oz[oz['likeCount'] > oz[['likeCount']].quantile(q = 0.95).values[0]]
plot_trend(ozMostLikes, fetMostLikes )
```

```python
plot_trend(ozMostLikes, fetMostLikes, BERT = True)
```

# Word Clouds

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image

def create_wordcloud(text):
    #mask = np.array(Image.open("cloud.png"))
    text = text.copy()
    
    not_include_words = ['john fetterman','John Fetterman','JOHN FETTERMAN','Fetterman','fetterman','John','john','dr Oz','Dr Oz',
                          'dr oz','DR OZ','OZ', 'oz','Oz','Dr.Oz','DR.OZ','dr.oz','Dr. Oz','DR. OZ','dr. oz', 'Mehmet',
                        'https', 'DrOz']
    f = lambda x: ' '.join(w for w in x.split() if not w in not_include_words)
    g = lambda x: ' '.join(w for w in x.split() if not w.startswith('@'))
    
    text['content'] = text['content'].apply(f)
    text['content'] = text['content'].apply(g)
    text_f= text['content']
    
    stopwords = set(STOPWORDS)
    wc = WordCloud(background_color="white",
                  #mask = mask,
                  max_words=3000,
                  stopwords=stopwords,
                  repeat=True)
    
    wc.generate(str(text_f))
    wc.to_file("wc.png")
    print("Word Cloud Saved Successfully")
    path="wc.png"
    display(Image.open(path))
```
