```python
import os
import pandas as pd
import datetime
import numpy as np
```

```python
initialDate = datetime.datetime(2022, 8, 1)
finalDate = datetime.datetime.today()

initialDateStr = str(initialDate).split(' ')[0]
finalDateStr = str(finalDate).split(' ')[0]

keywords = ['Fetterman', "'Dr Oz'", "'Mehmet Oz'"]
```

```python
def constructQuery(keyword, startDate, finalDate, nameFile):
    message = 'snscrape --jsonl twitter-search "{} since:{} until:{}" > {}.json'.format(keyword, startDate, finalDate, nameFile)
    return message

def constructDates(initialDate, finalDate):

    delta = (finalDate - initialDate).days + 1
    dates = [str(initialDate + datetime.timedelta(days=x)).split(' ')[0] for x in range(delta)]
    
    return dates
```

```python
dates = constructDates(initialDate, finalDate)
os_response = np.zeros(shape = (len(keywords), len(dates)))
```

```python
%%time

for j, keyword in enumerate(keywords):
    
    for i in range(1, len(dates)):

        start = dates[i - 1]
        final = dates[i]
            
            
        if ("'Dr Oz'" == keyword):
            name = "Dr_Oz"
            nameFile = 'Data/{}_{}'.format(name, start)
            
        elif ("'Mehmet Oz'" == keyword):
            name = "Mehmet_Oz"
            nameFile = 'Data/{}_{}'.format(name, start)
            
        else:
            nameFile = 'Data/{}_{}'.format(keyword, start)
            
            
        #Get the data
        message = constructQuery(keyword = keyword, startDate = start, finalDate = final, nameFile = nameFile)
        
        print(message)
        os_response[j, i] = os.system(message)
        
print(os_response.sum())
```
